```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Sagar Joshi\Downloads\Python_Diwali_Sales_Analysis\Python_Diwali_Sales_Analysis\Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
# GIVES INFORMATION ABOUT DATA
df.info()
```

```python
# PERMANENTLY DELETE NULL COLUMNS USING INPLACE FUNCTION
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
# SHOW DELETED COLUMNS
df.head()
```

```python
# FIND NULL ROWS
pd.isnull(df).sum()
```

```python
# PERMANENTLY DELETED NULL ROWS
df.dropna(inplace=True)
```

```python
# SHOW NULL ROWS
pd.isnull(df).sum()
```

```python
# CHANGE DATA TYPE
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
# SHOW HOW MANY COLUMNS 
df.columns
```

```python
# RENAME COLUMN 
df.rename(columns={'Marital_Status':'Marital'})
```

```python
# DO MATHEMATICAL CALCULATION
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
# PLOT A GRAPH
sns.countplot(df,x='Gender',palette='brg')
plt.show()
```

```python
# PLOT A GRAPH WITH VALUES
a=sns.countplot(df, x='Gender',palette='prism')

for bars in a.containers:
   a.bar_label(bars)
plt.show()
```

```python
# USING GROUP BY CLAUSE ON THE BASIS OF GENDER
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# PLOT A GRAPH FOR GROUP BY CLAUSE
b=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(b,x='Gender',y='Amount',palette='prism')
plt.show()
```

```python
df.columns
```

```python
# PLOT A GRAPH USING AGE AND GENDER
c=sns.countplot(df,x='Age Group', hue='Gender',palette='prism')
for bars in c.containers:
    c.bar_label(bars)
```

```python
# CREATE A GROUP BY CLAUSE BY AGE GROUP AND PLOT A GRAPH
d=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(d,x='Age Group',y='Amount',palette='prism')
plt.show()
```

```python
# PLOT A GRAPH OF ORDERS FROM TOP 10 STATES
e= df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(e,x='Orders',y='State',palette='prism') 
plt.show()
```

```python
# PLOT A GRAPH OF AMOUNT FROM TOP 10 STATES
e.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,10)})
sns.barplot(e,x='State',y='Amount',palette='prism') 
plt.show()
```
